#### #Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.
##### #Ans.1. Descriptive Statistics:
Definition:\
Descriptive statistics involves methods for organizing, summarizing and displaying data in a meaningful way. It deals only with the data that is actually collected.

Purpose:\
To describe the basic features of the data and provide simple summaries about the sample and the measures.

Techniques Used:\
Measures of central tendency: Mean, Median, Mode\
Measures of dispersion: Range, Variance, Standard Deviation\
Graphical representations: Bar charts, Pie charts, Histograms

Example:\
If the marks of 100 students in a class are collected, and the average (mean) score is calculated as 72%, highest score as 95%, and standard deviation as 8, then these values describe the performance of that class.

2.Inferential Statistics:\
Definition:\
Inferential statistics involves methods that use data from a sample to make conclusions or predictions about a larger population.

Purpose:\
To draw inferences, test hypotheses, and make predictions beyond the immediate data.

Techniques Used:\
Hypothesis Testing (t-test, z-test, chi-square test)\
Confidence Intervals\
Correlation and Regression Analysis

Example:\
Suppose a survey is conducted on 200 college students out of a total of 5,000. If 60% of them prefer online learning, inferential statistics helps in estimating that approximately 60% ± margin of error of all 5,000 students might prefer online learning.



#### #Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.
##### #Ans.Sampling is the process of selecting a small group or subset from a larger population in order to study it and draw conclusions about the entire population.It is commonly used when studying the whole population is not practical due to time, cost, or resource limitations.Sampling helps in making reliable estimates about the population using data from a smaller group.

Differences between Random Sampling and Stratified Sampling:

1.Basis of Selection:\
Random Sampling: The sample is selected from the entire population without dividing it into groups.\
Stratified Sampling: The population is first divided into groups (strata), and then samples are taken from each group.

2.Chance of Selection:\
Random Sampling: Every individual has an equal chance of being selected.\
Stratified Sampling: Every individual within each stratum has an equal chance of being selected.

3.Subgroup Representation:\
Random Sampling: May not guarantee representation of all subgroups in the sample.\
Stratified Sampling: Ensures all key subgroups of the population are well represented.

4.Type of Population:\
Random Sampling: Suitable when the population is homogeneous (similar in nature).\
Stratified Sampling: Suitable when the population is heterogeneous (diverse or mixed).

5.Example:\
Random Sampling: Selecting 100 students randomly from a college of 1000 students.\
Stratified Sampling: Selecting 60 boys and 40 girls from a college of 600 boys and 400 girls.



#### #Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.
##### #Ans.
Mean:\
Mean is the average of a set of values. It is calculated by adding all the values and dividing the sum by the total number of values.\
Example: For the numbers 10, 20, 30\
Mean = (10 + 20 + 30) ÷ 3 = 60 ÷ 3 = 20

Median:\
Median is the middle value of a data set when the values are arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values.\
Example: For the numbers 5, 10, 15\
Median = 10\
For the numbers 5, 10, 15, 20\
Median = (10 + 15) ÷ 2 = 12.5

Mode:\
Mode is the value that occurs most frequently in a data set. A data set may have no mode, one mode, or more than one mode.\
Example: For the numbers 4, 5, 5, 6
Mode = 5

Importance of Mean, Median, and Mode:

1.They help to summarize a large amount of data using a single value.\
2.They give a central idea or average of the data collected.\
3.They are used in various fields like economics, business, education and research.\
4.They help in comparing different sets of data.\
5.Median is useful when data contains extreme values or outliers.


#### #Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?
##### #Ans.Skewness:
Skewness is a statistical measure that tells us about the asymmetry of a data distribution.
If the data is symmetrical, skewness is zero. If it is not symmetrical, it is either positively skewed or negatively skewed.

Kurtosis:\
Kurtosis is a measure of the shape of a data distribution, especially the height and sharpness of the peak and the heaviness of the tails.\
It shows whether the data has more or fewer outliers compared to a normal distribution.

A positive skew means the right tail of the distribution is longer than the left tail.
Most of the data values are concentrated on the left, and a few large values pull the mean to the right.
In a positively skewed distribution, the mean is greater than the median, and the median is greater than the mode.
(Mean > Median > Mode)

Example:
Income data is often positively skewed because most people earn average incomes, while a few earn very high incomes.

```python
#Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
#(Include your Python code and output in the code box below.)

import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculating mean, median, and mode
mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean  :", mean)
print("Median:", median)
print("Mode  :", mode)
```

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
# list_x = [10, 20, 30, 40, 50]
# list_y = [15, 25, 35, 45, 60]
# (Include your Python code and output in the code box below.)

import numpy as np


list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

x = np.array(list_x)
y = np.array(list_y)

# Calculate covariance matrix
cov_matrix = np.cov(x, y, bias=False)  # bias=False uses sample covariance
covariance = cov_matrix[0][1]

# Calculate correlation coefficient
correlation = np.corrcoef(x, y)[0][1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
# Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
# data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
# (Include your Python code and output in the code box below.)

import matplotlib.pyplot as plt
import numpy as np

# Given dataset
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Plotting the boxplot
plt.figure(figsize=(6, 4))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of Data")
plt.ylabel("Value")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Calculate Q1, Q3 and IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Define bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers
outliers = [value for value in data if value < lower_bound or value > upper_bound]
print("Outliers:", outliers)



```

```python
# Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
# ● Explain how you would use covariance and correlation to explore this
# relationship.
# ● Write Python code to compute the correlation between the two lists:
# advertising_spend = [200, 250, 300, 400, 500]
# daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]


x = np.array(advertising_spend)
y = np.array(daily_sales)

cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0][1]

# Compute correlation
correlation = np.corrcoef(x, y)[0][1]

# Print results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
# Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
# ● Explain which summary statistics and visualizations (e.g. mean, standard
# deviation, histogram) you’d use.
# ● Write Python code to create a histogram using Matplotlib for the survey data:
# survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

import matplotlib.pyplot as plt
import numpy as np

# Customer survey data (1–10 scale)
scores = np.array([7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7])

# Plot histogram
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(scores, bins=range(4, 12), color='lightcoral', edgecolor='black', alpha=0.8)

# Customize plot
ax.set_title("Distribution of Customer Satisfaction Scores", fontsize=14)
ax.set_xlabel("Satisfaction Score")
ax.set_ylabel("Number of Customers")
ax.grid(visible=True, linestyle=':', linewidth=0.7, alpha=0.6)

plt.tight_layout()
plt.show()
```
